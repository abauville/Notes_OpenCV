# Line detection

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
import ipywidgets as widgets
```

```python
img_out = []
@interact(param1=widgets.IntSlider(min=3, max=20, step=2, value=0),
          param2=widgets.IntSlider(min=1, max=20, step=1, value=0))
def threshold(param1, param2):
    img = cv2.imread("test_img.jpg")
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = cv2.GaussianBlur(img, (5,5),100)
    plt.clf()
    img = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, param1, param2)
    plt.imshow(img,vmin=0,vmax=255,cmap='gray')
    global img_out
    img_out = img
```

```python
img2 = img_out.copy()

img = cv2.GaussianBlur(img2, (5,5),0)
img2 = cv2.Canny(img2, 10, 30)
plt.clf()
plt.imshow(img2,vmin=0,vmax=255,cmap='gray')
```
